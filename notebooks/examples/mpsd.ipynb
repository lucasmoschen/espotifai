## Use case: example

I want the most agressive Epica playlist.

```python
epica = df[(df.artist_name == 'Epica') & (~pd.isna(df.mood_aggressive_prob))]
groupby = epica.groupby('playlist_id').mean()
groupby.sort_values(by='mood_aggressive_prob', ascending=False)
```
